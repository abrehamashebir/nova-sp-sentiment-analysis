```python
import matplotlib.pyplot as plt

def plot_price_lines(df, ticker):
    plt.figure(figsize=(20, 8))
    
    plt.plot(df['Date'], df['Close'], label='Close')


    plt.title(f"{ticker} - Stock Price Overview")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for ticker in tickers:
    plot_price_lines(df, ticker)
```
